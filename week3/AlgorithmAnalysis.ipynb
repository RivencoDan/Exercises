### Algorithm Analysis

#### Complexity measures

1. Time Complexity: number of basic operations an algorithm performs, as a function of input size $n$. $\newline$
2. Space Complexity: extra memory used (beyond input), also as a function of $n$. $\newline$
3. Cost Model: what you count as a “basic operation” (e.g. comparisons, arithmetic ops).
